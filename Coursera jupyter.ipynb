```python
import yfinance as yf

t = yf.Ticker("TSLA")
td = t.history(period="max")
td.reset_index(inplace=True)
td.head()
```

```python
td.tail()
```

```python
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
gme_data.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(td['Date'], td['Close'], color='blue', label='Precio de Cierre')
plt.title('Tesla Stock Price Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Stock Price (USD)', fontsize=14)
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(gme_data['Date'], gme_data['Open'], label='Open', color='orange', linewidth=2, linestyle='--')
plt.plot(gme_data['Date'], gme_data['High'], label='High', color='green', linewidth=2)
plt.plot(gme_data['Date'], gme_data['Low'], label='Low', color='red', linewidth=2)
plt.plot(gme_data['Date'], gme_data['Close'], label='Close', color='blue', linewidth=2)
plt.title('GameStop Stock Data (Open, High, Low, Close)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Stock Price (USD)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
